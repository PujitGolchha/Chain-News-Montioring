```python
import requests
import numpy as np
import pandas as pd
from matplotlib import legend
import plotly.express as px
```

```python
country_counts = requests.get('https://8970-2001-7c0-2900-8050-5652-ff-fe00-18.eu.ngrok.io/locations/?limit=100').json()
```

```python
len(country_counts)
```

```python
country_counts_df = pd.DataFrame(country_counts)
country_counts_df
country_df = country_counts_df.groupby(['State','Country',  'Longitude', 'Latitude'], as_index=False)['ID'].count()
```

```python
country_counts_df.groupby(['RSS'], as_index=False)['ID'].count().sort_values('ID', ascending=False)
```

```python
len(country_df['Country'].unique())
```

```python
country_df.info()
```

```python
fig = px.scatter_mapbox(country_df, lat = 'Latitude', lon = 'Longitude', size = 'ID',
                        color = 'Country', color_continuous_scale = 'plasma',
                        zoom = 1, mapbox_style = 'carto-darkmatter', width=1200, height=800)
fig.update_layout(showlegend=False)

fig.show()
```
