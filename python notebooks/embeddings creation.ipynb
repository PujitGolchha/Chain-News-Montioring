```python
import gensim
from gensim.models import KeyedVectors
from gensim import models
import pandas as pd
from tqdm import tqdm
import numpy as np
import re
from sklearn.metrics import confusion_matrix
import pickle
from os import listdir
from os.path import isfile, join
import csv
```

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
stop_words = list(stopwords.words('english'))
```

```python
manual_input=["say","worker","company","said"]
stop_words=stop_words+manual_input
```

# Embeddings creation

```python
w2v_model = models.KeyedVectors.load_word2vec_format('/content/drive/MyDrive/Chain_News_Monitoring_Project/Pujit/GoogleNews-vectors-negative300.bin.gz', binary=True)
```

```python
path="/content/drive/MyDrive/Chain_News_Monitoring_Project/Paula"
onlyfiles = [join(path, f) for f in listdir(path) if isfile(join(path, f))]
onlyfiles = [f for f in onlyfiles if "list_of_dictionary" in f]
```

```python
data=[]
for i in onlyfiles:
  with open(i, "rb") as jfile:
    loaded_links=pickle.load(jfile)
  data=data+loaded_links

data=pd.DataFrame(data)
```

```python
data.to_csv("/content/drive/MyDrive/Chain_News_Monitoring_Project/Pujit/testing_data.csv",index=False)
```

```python
def document_vector(doc):
    # remove out-of-vocabulary words
      doc = [word for word in doc if word in w2v_model.wv.vocab]
      return np.mean(w2v_model[doc], axis=0)
```

```python
def preprocessing(document):
    # Remove punctuation and numbers
    try:
        
      document = re.sub(r"[^A-Za-z\u00c0-\u017e\s]+", " ", document)

      # Remove extra white spaces
      doc= re.sub(r"[\s]{2,}", " ", document)
      
      doc = doc.lower()
      doc = word_tokenize(doc)
      
      doc = [lemmatizer.lemmatize(word) for word in doc]
      doc = [word for word in doc if word not in stop_words]
    except:
      doc=[]
    return doc
```

```python
[preprocessing(title) for title in tqdm(list(data["title"][0:10]))]
```

```python
corpus_title = [preprocessing(title) for title in tqdm(list(data["title"]))]      
```

```python
corpus = [preprocessing(title) for title in tqdm(list(data["content"]))]
```

```python
header=["id","ts"]+list(range(0, 300))
header
```

```python
with open('/content/drive/MyDrive/Chain_News_Monitoring_Project/Pujit/test_w_2_vec.csv', 'w', encoding='UTF8') as f:
    writer = csv.writer(f)

    # write the header
    writer.writerow(header)

    
    for i,j,k,l in tqdm(zip(corpus,data["id"],data["ts"],corpus_title)):
      finaldf=[]
      try:
        doc_Embedding=document_vector(i)
        finaldf= finaldf+[j]+[k]+list(doc_Embedding)
        writer.writerow(finaldf)
      except:
        try:
          doc_Embedding=document_vector(l)
          finaldf= finaldf+[j]+[k]+list(doc_Embedding)
          writer.writerow(finaldf)
        except:
          pass
      
```

```python
finaldf=pd.DataFrame(finaldf)
finaldf[list(range(0, 300))] = pd.DataFrame(finaldf.emebedding.tolist(), index= finaldf.index)
finaldf.drop(columns='emebedding', inplace= True)
finaldf
```

```python
finaldf.sort_values(by=["ts"],inplace=True)
```

```python
finaldf.drop_duplicates(subset=["id","ts"],inplace=True) # accounting for none ts 
```
